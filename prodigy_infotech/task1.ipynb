```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("t1.csv", skiprows=4)
df = df[["Country Name", "Country Code", "2024"]]
df = df.dropna()
df["2024"] = pd.to_numeric(df["2024"], errors="coerce")
df_sorted = df.sort_values(by="2024", ascending=False)
top10 = df_sorted.head(10)
plt.figure(figsize=(10,6))
plt.bar(top10["Country Name"], top10["2024"], color="skyblue", edgecolor="black")
plt.xticks(rotation=45, ha="right")
plt.title("Top 10 Most Populous Countries (2024)")
plt.xlabel("Country")
plt.ylabel("Population (in billions)")
plt.tight_layout()
plt.show()
```

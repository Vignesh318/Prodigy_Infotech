```python
import pandas as pd 
import matplotlib.pyplot as plt

d=pd.read_csv("amazon_sales_dataset.csv")
print(d.head(1))
print(d.isnull().sum())


#top5
category_sales = d.groupby("product_category")["quantity_sold"].sum()

category_sales.plot(kind='bar')

plt.title("Top Selling Product Categories")
plt.xlabel("Product Category")
plt.ylabel("Quantity Sold")

plt.show()

plt.scatter(d["discount_percent"], d["quantity_sold"])

plt.title("Discount vs Quantity Sold")
plt.xlabel("Discount Percent")
plt.ylabel("Quantity Sold")

plt.show()
```
